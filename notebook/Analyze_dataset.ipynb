```python
with open(workdir / 'data/input/Asteria_Eng_ver5.pickle', 'rb') as f:
    data = pickle.load(f)
```

```python
seqlen_dic = defaultdict(list)
for uid, seq in data.items():
    last_learned_sid = -1
    sid_len = 0
    for qa in seq:
        sid = qa[0]
        if sid == last_learned_sid:
            sid_len += 1
        elif sid != last_learned_sid and last_learned_sid == -1:
            last_learned_sid = sid
            continue
        else:
            seqlen_dic[sid].append(sid_len)
            sid_len = 0
            last_learned_sid = sid
```

```python
ax = plt.subplot(111)
ax.set_xlim(0., 50.)
for s, sl in seqlen_dic.items():
    sns.distplot(sl)
```
