```python
def analyze_qa_format_data(data: List[Tuple[int, int]]):
    print('Students num:', len(data))
    
    print('Students exercise 0-9 mean dist across students:')
    sns.distplot([mean([a for q, a in s[:10]]) for s in data])
    plt.show()
    print('Students exercise 10-19 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[10:20]]) for s in data])
    plt.show()
    print('Students exercise 20-29 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[20:30]]) for s in data])
    plt.show()
    print('Students exercise 30-39 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[30:40]]) for s in data])
    plt.show()
    print('Students exercise 40-49 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[40:50]]) for s in data])
    plt.show()
    
    print('Students exercise 0-49 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s]) for s in data])
    plt.show()
    
    print('Students mean:')
    sns.distplot([ndcg_score(np.asarray([[i for i in range(10)]]), np.asarray([[a for q, a in s[:10]]])) for s in data])
    plt.show()
```

```python
def compare_qa_format_data(data1: List[Tuple[int, int]], data2):
    print('Students num:', len(data1))
    print('Students num:', len(data2))
    
    print('Students exercise 0-9 mean dist across students:')
    sns.distplot([mean([a for q, a in s[:10]]) for s in data1])
    sns.distplot([mean([a for q, a in s[:10]]) for s in data2])
    plt.show()
    print('Students exercise 10-19 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[10:20]]) for s in data1])
    sns.distplot([mean([a for q, a in s[10:20]]) for s in data2])
    plt.show()
    print('Students exercise 20-29 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[20:30]]) for s in data1])
    sns.distplot([mean([a for q, a in s[20:30]]) for s in data2])
    plt.show()
    print('Students exercise 30-39 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[30:40]]) for s in data1])
    sns.distplot([mean([a for q, a in s[30:40]]) for s in data2])
    plt.show()
    print('Students exercise 40-49 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s[40:50]]) for s in data1])
    sns.distplot([mean([a for q, a in s[40:50]]) for s in data2])
    plt.show()
    
    print('Students exercise 0-49 mean dist across studnts:')
    sns.distplot([mean([a for q, a in s]) for s in data1])
    sns.distplot([mean([a for q, a in s]) for s in data2])
    plt.show()
    
    print('Students learning ability 0-9:')
    sns.distplot([ndcg_score(np.asarray([[i for i in range(10)]]), np.asarray([[a for q, a in s[:10]]])) for s in data1])
    sns.distplot([ndcg_score(np.asarray([[i for i in range(10)]]), np.asarray([[a for q, a in s[:10]]])) for s in data2])
    plt.show()
    print('Students learning ability 0-49:')
    sns.distplot([ndcg_score(np.asarray([[i for i in range(50)]]), np.asarray([[a for q, a in s]])) for s in data1])
    sns.distplot([ndcg_score(np.asarray([[i for i in range(50)]]), np.asarray([[a for q, a in s]])) for s in data2])
    plt.show()
```

```python
compare_qa_format_data(synthetic, reproduction_v4)
```
